```python
# Connect to the database
conn = sqlite3.connect('arXiv.db')

# Query all values from the papers table
query = 'SELECT * FROM articles'
cursor = conn.cursor()
cursor.execute(query)
data = cursor.fetchall()

# Get the column names from the cursor's description
column_names = [desc[0] for desc in cursor.description]

# Create a DataFrame from the queried data
df = pd.DataFrame(data, columns=column_names)

# Close the connection
conn.close()

# Print the DataFrame
df
```

# Alex API Test Using DOI

```python
import json
import requests
import random

random_index = random.randint(0, len(df))

# URL of the JSON data
base_url = "https://api.openalex.org/works/https://doi.org/"
url = base_url + df.iloc[random_index]['doi']

# Fetch the JSON data
response = requests.get(url)
data = response.json()

# Parse authorships
title = data.get("title", "")
print('Title:'+'\n')
print(title)
print('---')
concepts = data.get("concepts", [])
print('Concepts:'+'\n')
for concept in concepts:
#     concept_id = concept.get("id", "")
    concept_name = concept.get("display_name", "")
    print(f"{concept_name}")
print('---')
print('Authors:'+'\n')
for authorship in data.get("authorships", []):
    author_name = authorship["author"].get("display_name", "")
    try:
        affiliation = authorship["institutions"][0].get("display_name", "")
    except:
        pass
    print(f"{author_name}, {affiliation}")
```
